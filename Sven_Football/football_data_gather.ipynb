```python
wc_m_data = world_cup_matches_json['data']
wc_m_data

num = range(0,64)
# for i in num:
#     print(i)
match_start = []
group_name = []
home_team_name = []
home_team_short_code = []
away_team_name = []
away_team_short_code = []
stats_home_score = []
stats_away_score = []
stats_halftime_score = []
stats_fulltime_score = []
venue_name = []
venue_capacity = []
venue_city = []

    
for i in num:
    match_start.append(wc_m_data[i]['match_start'])
    group_name.append(wc_m_data[i]['group']['group_name'])
    home_team_name.append(wc_m_data[i]['home_team']['name'])
    home_team_short_code.append(wc_m_data[i]['home_team']['short_code'])
    away_team_name.append(wc_m_data[i]['away_team']['name'])
    away_team_short_code.append(wc_m_data[i]['away_team']['short_code'])
    stats_home_score.append(wc_m_data[i]['stats']['home_score'])
    stats_away_score.append(wc_m_data[i]['stats']['away_score'])
    stats_halftime_score.append(wc_m_data[i]['stats']['ht_score'])
    stats_fulltime_score.append(wc_m_data[i]['stats']['ft_score'])
    venue_name.append(wc_m_data[i]['venue']['name'])
    venue_capacity.append(wc_m_data[i]['venue']['capacity'])
    venue_city.append(wc_m_data[i]['venue']['city'])
    
    world_cup_dataframe = pd.DataFrame({'match_start':match_start, 'group_name':group_name,
                       'home_team_name':home_team_name, 'home_team_short_code':home_team_short_code,
                       'away_team_name':away_team_name, 'away_team_short_code':away_team_short_code,
                       'stats_home_score':stats_home_score, 'stats_away_score':stats_away_score,
                       'stats_halftime_score':stats_halftime_score, 'stats_fulltime_score':stats_fulltime_score,
                       'venue_name':venue_name, 'venue_capacity':venue_capacity,
                       'venue_city':venue_city})

    

```

```python
world_cup_dataframe.to_csv('world_cup_dataframe.csv', index=False)
```
